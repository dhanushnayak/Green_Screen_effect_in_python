### Import Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
% matplotlib inline
```

### Read - Display Image

```python
# Read the image
img = cv2.imread("./images/car.jpg")


print('This is the image Type : ',type(img),'with dimensions: ',img.shape)
```

```python
# Make a copy of the image
img_copy = img.copy()

# Change the color from BGR to RGB
img_copy = cv2.cvtColor(img_copy,cv2.COLOR_BGR2RGB)



# Display the image
plt.imshow(img_copy)
```

## Define Color Threshold

```python

lower_green = np.array([0,200,0])
upper_green = np.array([250,255,200])
```

## Creat the Mask

```python
# Define the Masked Area
mask = cv2.inRange(img_copy,lower_green,upper_green)

# Visualize the Mask on the imgae
plt.imshow(mask,cmap='gray')
```

```python
# Mask the Image to let your object come and show through
masked_image = img_copy.copy()

masked_image[mask != 0] = [0,0,0]
# Dispaly the Image
plt.imshow(masked_image)
```

## Mask and Add a Background Image

```python
# Load the Image Background
background_image = cv2.imread("./images/london.jpg")

# Convert it from BGR to RGB
background_image = cv2.cvtColor(background_image,cv2.COLOR_BGR2RGB)

# Crop it to the the dimensions we got
crop =  background_image[0:450,0:660]

# Mask the Cropped Background
# That's how the object's area is blocked
crop[mask == 0] = [0,0,0]

# Display the Background
plt.imshow(crop)
```

## Let's move it

```python
# x, y, h, w
x=390
y=300
h=450
w=660

# Convert it from BGR to RGB
crop = cv2.imread("./images/london.jpg")
crop = cv2.cvtColor(crop,cv2.COLOR_BGR2RGB)

# Crop it to the the dimensions we got
crop = crop[y:y+h,x:x+w]

# Mask the Cropped Background
# That's how the object's area is blocked
crop[mask == 0] =[0,0,0]

# Display the Background
plt.imshow(crop)
```

# Add them together and get the Final Image

```python
complete_img = masked_image + crop
plt.imshow(complete_img)
```
